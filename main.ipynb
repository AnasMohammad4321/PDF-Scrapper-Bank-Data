```python
import pandas as pd
import numpy as np
import os
```

```python
csv_files = [file for file in os.listdir() if file.endswith("_withdrawals.csv")]

dataframes = []

for file in csv_files:
    df = pd.read_csv(file)
    dataframes.append(df)

df = pd.concat(dataframes, ignore_index=True)
```

```python
# # Create a new 'Category' column in the DataFrame
df["Category"] = "Others"

# Check and update category based on keywords
for category, keywords in dict.items():
    for keyword in keywords:
        mask = df["Description"].apply(lambda x: keyword in str(x))
        # Update 'Category' column where keyword is found
        df.loc[mask, "Category"] = category
```
